#### How To Change And Reset Index


```python
import pandas as pd
import numpy as np
#import csv file
my_df =pd.read_csv('dog_data_short.csv')
my_df
```

```python
# create new column
my_df["Frame Header"]=["Dog 1","Dog 2","Dog 3","Dog 4","Dog 5"]
my_df
```

```python
#set index
my_df.set_index("Frame Header",inplace=True)
```

```python
# show dataframe
my_df
```

```python
# reset index
my_df.reset_index()
my_df
```

```python
# reset index permenantly
my_df.reset_index(inplace=True)
my_df
```

```python
# drop column
my_df.drop("Frame Header", axis=1,inplace=True)
```
