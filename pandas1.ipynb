```python
import pandas as pd
```

### Basics

```python
# Pandas series like a table column - 1D arraty holding any type of data

my_series = [2,5,7,99,10]
my_var = pd.Series(my_series)
print(my_var)
```

```python
# print specific value

my_var[3]
```

```python
# labels using index augment
my_index= ['i','ii','iii','iv','v']
my_var2=pd.Series(my_series,index=my_index)
print(my_var2)
```

```python
print(my_var2['iii'])
print(my_var2[4])
```

```python
#key value dictionaries

cars= {"tesla":12,"mercedes":42,"bmw":3}
my_var3 = pd.Series(cars)
my_var3
```

```python
print(my_var3['tesla'])
```
