#### How To Make Conditional Selections


```python
import pandas as pd
import numpy as np
#import csv file
my_df =pd.read_csv('dog_data_short.csv')
my_df
```

```python
# Conditional Selection
# > < = < = etc
# Return Booleans

my_df =='BROWN'

```

```python
# return data frame with data
my_df[my_df =='BROWN']
```

```python
# run them on a column
my_df[my_df =='BROWN']['Color']
```

```python
# return multiple column
my_df[my_df =='BROWN'][['Color','Breed']]
```
