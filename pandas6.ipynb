#### Grab Rows, Points, and Subsets From DataFrame

```python
import pandas as pd
import numpy as np
#import csv file
my_df =pd.read_csv('dog_data_short.csv')
my_df
```

```python
# grab row using Location
my_df.loc[2] 
#NOTE: loc[] - contains name of row
```

```python
# grab row using index Location
my_df.iloc[3]
#NOTE: iloc[] - contains index of row
```

```python
#Specific Points (row and column)
#my_df.loc['Row', 'COLUMN']

my_df.loc[1,'DogName']
```

```python
#Subsets
#my_df.loc[['row', 'row'], ['column','column']]
my_df.loc[[3,1],['Breed','DogName']]
```
